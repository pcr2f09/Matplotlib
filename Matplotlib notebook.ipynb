```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
filename = os.path.join("Resources", "mouse_drug_data.csv")
filename = os.path.join("Resources", "city_data.csv")
drug_df = pd.read_csv(filename)
drug_df.head(15)
```

```python
x_axis = np.arange(len(drug_df))
tick_locations = [value+0.4 for value in x_axis]
```

```python
plt.figure(figsize=(20,3))
plt.bar(x_axis, drug_df["driver_count"], color='r', alpha=0.5, align="edge")
plt.xticks(tick_locations, drug_df["city"], rotation="vertical")
```

```python
plt.xlim(-0.25, len(x_axis))
plt.ylim(0, max(drug_df["driver_count"])+10)
```

```python
plt.title("Driver Count per State")
plt.xlabel("City")
plt.ylabel("Driver Count")
```

```python
plt.tight_layout()
plt.savefig("avg_driver_count.png")
plt.show()
```
